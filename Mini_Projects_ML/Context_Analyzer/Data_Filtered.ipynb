```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp=spacy.load("en_core_web_md")
```

```python
stopword_list=[STOP_WORDS]
```

```python
reviews_list=[]
j=0
for Each_review in df.reviews:
        temp_list=str(Each_review)
        reviews_list.append([temp_list])
```

```python
len(reviews_list)
```

```python
reviews_list[0]
```

```python
import spacy_cleaner
from spacy_cleaner.processing import removers
```

```python
pipeline=spacy_cleaner.Pipeline(
nlp,
removers.remove_punctuation_token,
removers.remove_stopword_token,
)
```

```python
for index in range(0,len(reviews_list)):
    reviews_list[index]=pipeline.clean(reviews_list[index])
```

```python
reviews_list[0]
```

```python
filtered_list=[]
j=0
for i in reviews_list:
    for j in i:
        str1=''.join(j)
    filtered_list.append(str1)
```
