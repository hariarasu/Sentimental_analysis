```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import spacy_cleaner
from spacy_cleaner.processing import removers
```

```python
reviews_list1 = []
for review in df2.text:
    # Split the review into tokens
    tokens = review.split()
    # Remove any tokens that start with "@"
    tokens = [token for token in tokens if not token.startswith("@")]
    # Rejoin the remaining tokens into a string
    cleaned_review = " ".join(tokens)
    reviews_list1.append([cleaned_review])
```

```python
len(reviews_list1)
```

```python
pipeline=spacy_cleaner.Pipeline(
nlp,
removers.remove_punctuation_token,
removers.remove_stopword_token,
)
```

```python
for index in range(0,len(reviews_list1)):
    reviews_list1[index]=pipeline.clean(reviews_list1[index])
```
